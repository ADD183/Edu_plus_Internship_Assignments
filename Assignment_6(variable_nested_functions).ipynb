```python
# 3 Write a program where a global variable is updated by multiple function calls using the global keyword

bal=2400

def income():
    global bal
    bal += 5000
    print("Salary added: ",bal)

def bills():
    global bal
    bal-=4000
    print("Bills subtracted:",bal)

income()
bills()

print("Total Balance:",bal)


```

```python
# 4 Create two functions where both use a global variable to calculate and store cumulative total sales
total_sales = 0

def add_online_sales():
    global total_sales
    online_sales = 1500
    total_sales = total_sales + online_sales
    print("Online Sales Added:", online_sales)
    print("Total Sales:", total_sales)

def add_offline_sales():
    global total_sales
    offline_sales = 2300
    total_sales = total_sales + offline_sales
    print("Offline Sales Added:", offline_sales)
    print("Total Sales:", total_sales)

add_online_sales()
add_offline_sales()

print("Final Cumulative Total Sales:", total_sales)


```

```python
# 6 Demonstrate the use of nonlocal keyword by modifying an enclosing variable inside an inner function
def sales_tracker():
    total_sales = 200   # Enclosing variable

    def add_sales():
        nonlocal total_sales   # Refers to enclosing function variable
        total_sales = total_sales + 100
        print("Sales Added. Total Sales:", total_sales)

    add_sales()

sales_tracker()
```

```python
# 13 Demonstrate how enclosing variable stores the running total for a counter function using nonlocal
def counter():
    count = 0   # Enclosing variable

    def increment():
        nonlocal count  
        count = count + 1
        print("Current Count:", count)

    increment()
    increment()
    increment()

counter()
```

```python
# 24 Demonstrate how a nonlocal variable can be used to maintain the count of how many times a nested function has been called
def call_counter():
    count = 0   # Enclosing variable

    def inner_function():
        nonlocal count
        count = count + 1
        print("Inner function called", count, "times")

    inner_function()
    inner_function()
    inner_function()

# Function call
call_counter()
```

```python
# 26 Create a nested function example where inner function modifies enclosing variable to accumulate running total of expenses
def expense_tracker():
    total_expense = 0   # Enclosing variable

    def add_expense(amount):
        nonlocal total_expense
        total_expense = total_expense + amount
        print("Added Expense:", amount)
        print("Total Expense:", total_expense)

    add_expense(1200)
    add_expense(800)
    add_expense(500)

# Function call
expense_tracker()
```

```python
# 27 Write a function that uses *args to concatenate multiple strings into a single string
def concatenate_strings(*args):
    result = ""
    for text in args:
        result = result + text
    return result

# Function calls
print(concatenate_strings("Hello", " ", "World"))
print(concatenate_strings("Python", " ", "is", " ", "awesome"))
```

```python
# 29 Write a function with both *args and **kwargs to simulate order details containing items and their prices
def order_details(*items, **prices):
    print("Order Items:")
    for item in items:
        print("-", item)

    print("\nItem Prices:")
    total = 0
    for key, value in prices.items():
        print(f"{key}:{value}")
        total = total + value

    print("\nTotal Amount:", total)

order_details("Laptop", "Mouse", "Keyboard",Laptop=55000, Mouse=800, Keyboard=1200)
```

```python
# 30 Design a nested function where the outer function provides base salary and the inner function adds incentives using nonlocal variable
def salary_structure(base_salary):
    total_salary = base_salary   # Enclosing variable

    def add_incentive(incentive):
        nonlocal total_salary
        total_salary = total_salary + incentive
        print("Added Incentive:", incentive)
        print("Total Salary:", total_salary)

    add_incentive(5000)
    add_incentive(3000)

# Function call
salary_structure(40000)
```
