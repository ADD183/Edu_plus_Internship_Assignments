```python
# 30. Create two tuples — (product_name, price) — and combine them using zip(); then display them sorted by price in descending order.
p_name=("Laptop","Smartphone","Tablet","Smartwatch")
prices=(1000, 800, 600, 400)
product_prices=tuple(zip(p_name, prices))
sor=sorted(product_prices, key=lambda x: x[1], reverse=True)
print(sor)   #30
```
