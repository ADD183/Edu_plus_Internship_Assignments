```python
'''Scenario

A software is being developed to help students and programmers for converting Hexadecimal numbers into 
Binary format.
Since user input can be inconsistent (extra spaces, lowercase letters, invalid characters, prefixes like 0x), 
the program must clean, validate, and correctly convert the input while ensuring proper formatting and correctness.'''
```

```python
'''Problem Statement

Write a Python program that accepts a hexadecimal number from the user and performs the following operations:'''
```

```python
#Input Cleaning and Formatting
Hex_input=input("Enter a Hexadecimal number: ")

#strip(): Removes extra spaces from both the beginning and end of a string.
clean_hex=Hex_input.strip()

#Prefix Handling
#replace():Replaces a specified string with another specified string.
clean_hex=clean_hex.replace("0x"," ")
clean_hex=clean_hex.replace("x0"," ")

#lstrip():Removes unwanted spaces only from the left side of a string.
clean_hex=clean_hex.lstrip()

#rstrip():Removes unwanted spaces only from the right side of a string.
clean_hex=clean_hex.rstrip()

#upper(): Converts all lowercase letters in a string into uppercase.
clean_hex=clean_hex.upper()

print("Cleaned Hexadecimal number is:",clean_hex)
```

```python
#Hexadecimal Validation
#isdidgit(): Checks whether a character is a digit (0-9).
for char in clean_hex:
    if not (char.isdigit() or (char>="A" and char<="F") or char=="."):
        print("Invalid Hexadecimal number")
        exit()
print("Valid Hexadecimal number")
```

```python
#Hexadecimal to Binary Conversion
#replace():Replaces a specified string with another specified string.
clean_hex=clean_hex.replace("0","0000")
clean_hex=clean_hex.replace("1","0001")
clean_hex=clean_hex.replace("2","0010")
clean_hex=clean_hex.replace("3","0011")
clean_hex=clean_hex.replace("4","0100")
clean_hex=clean_hex.replace("5","0101")
clean_hex=clean_hex.replace("6","0110")
clean_hex=clean_hex.replace("7","0111")
clean_hex=clean_hex.replace("8","1000")
clean_hex=clean_hex.replace("9","1001")
clean_hex=clean_hex.replace("A","1010")
clean_hex=clean_hex.replace("B","1011")
clean_hex=clean_hex.replace("C","1100")
clean_hex=clean_hex.replace("D","1101")
clean_hex=clean_hex.replace("E","1110")
clean_hex=clean_hex.replace("F","1111")


print(clean_hex)
```

```python
#If the hexadecimal number contains a decimal point, split the number into:
#split() :Splits a string into a list where each word is a list item.
if "." in clean_hex:
    int_hex, fract_hex = clean_hex.split(".")
else:
    int_hex = clean_hex
    fract_hex = ""
print("Integer part in Binary:",int_hex)  #Printing only integer part of binary number

#join():Joins the elements of an iterable to the end of the string.
final_hex=int_hex+"." +"".join(fract_hex)
print("Binary number after conversion:",final_hex)  #Printing the binary number after conversion
```

```python
#Removal of Unnecessary Zeros
final_hex=final_hex.rstrip('0')
final_hex=final_hex.rstrip('.') 
final_hex=final_hex.lstrip('0')
```

```python
#Output Validation
#isalnum(): Checks whether all characters in a string are alphanumeric (either alphabets or numbers).
print("Is the hexadecimal number converted?",final_hex.isalnum())

#Final Output
print("Final Binary number is:",final_hex)
```

```python
#Counting Hexadecimal Digits
print("Final Binary number is:",final_hex)
hex_digits = 0
for ch in final_hex:
    if ch != ".":
        hex_digits += 1

print("Number of Binary digits:", hex_digits)
```

```python
#Finnding Addition of Binary Digits
sum_digits = 0  
for ch in final_hex:
    if ch != ".":
        sum_digits += int(ch)
print("Sum of Binary digits:", sum_digits)
```
