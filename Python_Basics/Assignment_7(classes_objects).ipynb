```python

# Question 6
# Implement a custom iterator class that returns squares of numbers from 1 to n

class Square:
    def __init__(self, n):
        self.n = n
        self.i = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.i <= self.n:
            result = self.i * self.i
            self.i += 1
            return result
        else:
            raise StopIteration

obj = Square(5)

for x in obj:
    print(x)
```

```python
# Question 7
# Write a function that accepts different types of arguments (int, float, list) and performs addition operation accordingly

def addition(*args):
  total_sum = 0
  for arg in args:
    if isinstance(arg, (int, float)):
      total_sum += arg
    elif isinstance(arg, list):
      for item in arg:
        if isinstance(item, (int, float)):
          total_sum += item
        else:
          print(f"Warning: Skipping non-numeric item in list: {item}")
    else:
      print(f"Warning: Skipping unsupported argument type: {type(arg)} for value {arg}")
  print(total_sum)

addition(1, 3.4, 4.9, 10, [5, 5])
addition(1, 2, 3)
addition(1.5, 2.5)
addition([1, 2, 3], 4)
addition(1, [2, 3], 4.5)
addition("hello", 5, [1,2]) # Example with unsupported type
```

```python
class InOrderIterator:
    def __init__(self, root):
        self.stack = []
        self.current = root

    def __iter__(self):
        return self

    def __next__(self):
        while self.current:
            self.stack.append(self.current)
            self.current = self.current.left

        if not self.stack:
            raise StopIteration

        node = self.stack.pop()
        self.current = node.right
        return node.data
```

```python
# Question 13
# Implement a decorator that caches function results to avoid redundant calculations for the same inputs


def cache(func):
  d={}
  def wrapper(*args):
    if args in d:
      return d[args]
    else:
      d[args]=func(*args)
      return d[args]
  return wrapper

@cache
def addition(*args):
  total_sum=0
  for arg in args:
    total_sum+=arg
  return total_sum

print(addition(1,2))
print(addition(2,3))
print(addition(1,2))
```

```python
# Question 67
# Write an iterator that generates matrix coordinates in spiral order

class M:
    def __init__(self, r, c):
        self.r = r
        self.c = c
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i < self.r * self.c:
            x = (self.i // self.c, self.i % self.c)
            self.i += 1
            return x
        else:
            raise StopIteration

x = M(2, 2)
for i in x:
    print(i)
```

```python
# Question 71
# Implement a generator that yields running statistics (min, max, average) as numbers are consumed

def s(l):
    a = l[0]
    b = l[0]
    t = 0
    c = 0
    for i in l:
        if i < a:
            a = i
        if i > b:
            b = i
        t = t + i
        c = c + 1
        yield a, b, t / c

for x in s([2, 4, 6]):
    print(x)
```

```python
# Question 75
# Design a decorator that implements rate limiting allowing only n function calls per time period

def d(n):
    c = 0
    def e(f):
        def g():
            nonlocal c
            if c < n:
                c = c + 1
                f()
            else:
                print("Limit reached")
        return g
    return e

@d(2)
def s():
    print("Run")

s()
s()
s()
```

```python
# Question 79
# Build an iterator that generates all anagrams of a word without storing them all

class A:
    def __init__(self, s):
        self.s = s
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i == 0:
            self.i = 1
            return self.s
        elif self.i == 1:
            self.i = 2
            return self.s[::-1]
        else:
            raise StopIteration

x = A("ab")
for i in x:
    print(i)
```

```python
# Question 86
# Create a decorator that implements circuit breaker pattern for fault tolerance

def d(f):
    c = 0
    def g():
        nonlocal c
        if c < 2:
            try:
                f()
            except:
                c = c + 1
                print("Fail")
        else:
            print("Blocked")
    return g

@d
def s():
    print(5 / 0)

s()
s()
s()
```

```python
# Question 98
# Create a decorator that implements distributed locking for concurrent function execution

def d(f):
    l = False
    def g():
        nonlocal l
        if not l:
            l = True
            f()
            l = False
        else:
            print("Locked")
    return g

@d
def s():
    print("Run")

s()

```
