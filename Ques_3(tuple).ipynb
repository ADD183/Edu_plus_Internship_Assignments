```python
marks=(85, 90, 78, 92, 88, 85, 95, 85)
print("Max:",max(marks))        #7

print("Min:",min(marks))        #7

print("Sum:",sum(marks))        #8

print(tuple(sorted(marks)))       #9

print(tuple(sorted(marks, reverse=True)))       #10

print(marks.count(85))          #11

print(marks.index(88))          #12 
```
