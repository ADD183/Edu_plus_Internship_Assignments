```python
# 3. Write a Python program to tokenize a given paragraph into sentences using sent_tokenize().

from nltk.tokenize import sent_tokenize

text="Rohit works as a software engineer at Google India. He lives in Pune and enjoys learning artificial intelligence. Google collaborates with Microsoft and OpenAI to build intelligent systems."

sentence=sent_tokenize(text)
print(sentence)
```

```python
# 4. Tokenize the same paragraph into individual words using word_tokenize().
from nltk.tokenize import word_tokenize

words=word_tokenize(text)
print(words)
```

```python
# 5. Remove stopwords from a given text using NLTK stopwords list.

from nltk.corpus import stopwords

stop_words=set(stopwords.words('english'))
filtered_words=[ w for w in words if w.lower() not in stop_words]

print("words after removing stopwords:\n",filtered_words)
```

```python
# 6. Perform stemming on a list of words using the PorterStemmer and print the output.

from nltk.stem import PorterStemmer

stemmer=PorterStemmer()
stemmed_words = [stemmer.stem(w) for w in filtered_words]

print("Stemmed words are:\n",stemmed_words)
```

```python
# 7. Perform lemmatization on the same list of words using WordNetLemmatizer and compare results with stemming.

from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(w) for w in filtered_words]

print(lemmatized_words)
```

```python
# 8. Apply Part of Speech (POS) tagging to a given sentence and print each word with its tag.

pos_tags=nltk.pos_tag(filtered_words)
print("Parts of Speech:\n",pos_tags)
```

```python
# 9. Perform frequency distribution on a paragraph and print the top 10 most frequent words.

from nltk import FreqDist

freq_dist=FreqDist(filtered_words)
print("words Frequency:\n")
for words,freq in freq_dist.most_common(10):
    print(f"{words}:{freq}")
```

```python
# 10. Plot a graph of the top 15 most frequent words using FreqDist().plot().

from nltk import FreqDist

freq_dist=FreqDist(filtered_words)
print("words Frequency:\n")
for words,freq in freq_dist.most_common(15):
    print(f"{words}:{freq}")

freq_dist.plot(15,title="Top 10 word Frequency")
```

```python
# 23. Create a function that checks whether specific words exist in a review text and returns True/False.

from nltk.corpus import movie_reviews
import nltk

words = movie_reviews.words()

def find_word(words):
    search = input("Enter the word to search: ")

    for w in words:
        if w == search:
            print(f"The Word {search} is present")
            return True

    print(f"The Word {search} is not present")
    return False


find_word(words)

```

```python
# 24. Prepare feature sets using the movie_reviews dataset and split them into training and testing data.

documents = []

for c in movie_reviews.categories():
    for f in movie_reviews.fileids(c):
        documents.append((list(movie_reviews.words(f)), c))

featuresets = []
for words, category in documents:
    d = {}
    for w in words:
        d[w] = True
    featuresets.append((d, category))

train_set = featuresets[:1500]
test_set = featuresets[1500:]

print("Training size:", len(train_set))
print("Testing size:", len(test_set))
```

```python
# 25. Train a Naive Bayes classifier using the movie_reviews dataset and print the accuracy.

from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy

classifier = NaiveBayesClassifier.train(train_set)

print("Accuracy:", accuracy(classifier, test_set))
```

```python
# 26. Display the top 10 most informative features of the Naive Bayes classifier.
classifier.show_most_informative_features(10)
```

```python
# 27. Write a function that takes a custom sentence and predicts if it is positive or negative using the trained model.

from nltk import word_tokenize

sentence = "This movie was very interesting and enjoyable"
words = word_tokenize(sentence)

features = {}
for w in words:
    features[w] = True

print("Sentiment:", classifier.classify(features))
```

```python
# 28. Save the trained Naive Bayes classifier using pickle and load it again to make predictions.
import pickle

# Save
f = open("model.pkl", "wb")
pickle.dump(classifier, f)
f.close()

# Load
f = open("model.pkl", "rb")
loaded_model = pickle.load(f)
f.close()

print(loaded_model.classify(features))
```

```python
# 29. Perform text preprocessing (tokenization, stopword removal, lemmatization) on user input text.
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize

nltk.download('stopwords')
nltk.download('wordnet')

text = "Rohit is learning artificial intelligence"
words = word_tokenize(text)

words = [w for w in words if w.isalpha()]

stop = stopwords.words('english')
words = [w for w in words if w not in stop]

lm = WordNetLemmatizer()
words = [lm.lemmatize(w) for w in words]

print(words)
```

```python

# 30. Create a complete text classification pipeline using NLTK from raw text to sentiment prediction.
text = "The movie was boring and slow"

words = word_tokenize(text)
words = [w for w in words if w.isalpha()]

features = {}
for w in words:
    features[w] = True

print("Prediction:", loaded_model.classify(features))
```
