```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Q1: Create a simple line plot of y = 2x + 3 for x values ranging from 0 to 10.

x=np.linspace(0,10,100)
y=(2*x+3)

plt.plot(x,y,"r--")
plt.title("Simple Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
# Q2: Plot sin(x) and cos(x) on the same graph with different colors, line styles, and labels, and display a legend.

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1, linestyle="--",marker="o",label='sin(x)')
plt.plot(x,y2,linestyle="--",marker="x",label='cos(x)')

plt.title("Trigonometric Functions")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Q3: Create a bar chart showing the sales of four products: A, B, C, and D with random values.

products=['A','B','C','D']
sales=[100,200,150,250]

plt.bar(products,sales)

plt.title("Bar Chart")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
# Q4: Generate a scatter plot with random x and y values between 0 and 1, and use different colors for points.

x=np.random.rand(1,50)
y=np.random.rand(1,50)
color=np.random.rand(50)
size=1000*np.random.rand(50)

plt.scatter(x,y,c=color,s=size,alpha=0.5,cmap='viridis')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Q5: Plot a histogram for 1000 random numbers generated from a normal distribution.

data=np.random.rand(1000)

plt.hist(data,bins=10,color='red',edgecolor='black',alpha=0.75)
plt.title("Histogram")
plt.xlabel("Data")
plt.ylabel("Data")
plt.show()
```

```python
# Q6: Draw a pie chart representing the percentage of five fruits: Apple, Banana, Mango, Grapes, and Orange.
fruits=['Apple', 'Banana', 'Mango', 'Grapes','Orange']
per=[25,20,25,15,15]
color=['Red','Yellow','Blue','Green','Orange']

plt.pie(per,labels=fruits,colors=color,startangle=90,shadow=True,autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
# Q7: Create two subplots in one figure  one showing a sine wave and the other showing a cosine wave.

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x,y1,"r--")
plt.title("siz(x)")

plt.subplot(1,2,2)
plt.plot(x,y2,"b--")
plt.title("cos(x)")

plt.suptitle("Multiple Plots Example")
plt.tight_layout()
plt.show()
```

```python
# Q8: Customize a line plot by adding title, axis labels, grid lines, and saving it as 'line_plot.png'.

x=np.linspace(0,5,100)
y=np.exp(-x)

plt.plot(x,y,'m--')
plt.title("Exponential Function")
plt.xlabel("X")
plt.ylabel("exp(X)")

plt.savefig("Matplotlib-library-exponential_plot.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
# Q9: Create a 2x2 grid of subplots showing different plots: line, bar, scatter, and histogram.

fig,axs= plt.subplots(2,2,figsize=(10,8))

x=np.linspace(0,10,100)
axs[0,0].plot(x,np.sin(x),'r-')
axs[0,0].set_title("Sine Wave")

categories=['A','B','C','D']
values=[10,15,7,12]
axs[0,1].bar(categories,values,color='teal')
axs[0,1].set_title("Bar Chart")

data=np.random.rand(1000)
axs[1,0].hist(data,bins=25,color='blue',edgecolor='black')
axs[1,0].set_title("Histogram")

x=np.random.rand(50)
y=np.random.rand(50)
axs[1,1].scatter(x,y,color='black')
axs[1,1].set_title("Scatter Plot")

plt.suptitle("Matplotlib Dashboard Examples")
plt.tight_layout()

plt.savefig("Matplotlib-Dashboard.png",dpi=300,bbox_inches='tight')

plt.show()
```

```python
# Q10: Create a line plot for y = e^x for x between 0 and 5, and change the color, marker, and line style.

x=np.linspace(0,5,100)
y=np.exp(x)

plt.plot(x,y,color='Red',marker='o',linestyle='--')
plt.title("Exponential Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-label")
plt.show()
```
